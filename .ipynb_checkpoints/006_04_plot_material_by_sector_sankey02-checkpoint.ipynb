## Sankey plot

```python
# source data from https://plotly.com/python/sankey-diagram/

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 100,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = label_s,
      color = 'thistle'
    ),
    link = dict(
      source = source_s, # indices correspond to labels, eg A1, A2, A1, B1, ...
      target = target_s,
      value = value_s
  ))])

fig.update_layout(title_text="Material emissions distribution accross construction sector in Ontario, Canada (MTCO2e, Consumption 2018)", font_size=12)
fig.show()
```
